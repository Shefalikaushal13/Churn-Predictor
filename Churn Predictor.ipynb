```python
import numpy as np
import pandas as pd
from sklearn.preprocessing import StandardScaler
```

```python
df=pd.read_csv("7 churn.csv")
df.head()
```

```python
columns_to_keep = [
    'gender', 'SeniorCitizen', 'Partner', 'Dependents', 'PhoneService', 'MultipleLines',
    'InternetService', 'OnlineSecurity', 'DeviceProtection', 'TechSupport',
    'StreamingTV', 'StreamingMovies', 'PaperlessBilling', 'PaymentMethod', 'Contract',
    'TotalCharges', 'Churn','tenure', 'MonthlyCharges'
] #feature selection

df = df[columns_to_keep]
df.head()
```

```python
from sklearn.preprocessing import LabelEncoder, StandardScaler

# List of categorical columns
categorical_cols = [
    col for col in [
        'gender', 'Partner', 'Dependents', 'PhoneService', 'MultipleLines',
        'InternetService', 'OnlineSecurity', 'DeviceProtection',
        'TechSupport', 'StreamingTV', 'StreamingMovies', 'Contract',
        'PaperlessBilling', 'PaymentMethod'
    ] if col in df.columns  # Include only existing columns
]

```

```python
df
```

```python
binary_columns = ['Partner', 'Dependents', 'PhoneService', 'Churn','OnlineSecurity','DeviceProtection','TechSupport','StreamingTV','StreamingMovies'
                 ,'PaperlessBilling']
df[binary_columns] = df[binary_columns].replace({'Yes':1,'No':0})
df[['MultipleLines','Contract']] = df[['MultipleLines','Contract']].replace({'Yes':1,'No':0,'No phone service': 3,"Month-to-month":0,'One year':1,'Two year':2})
df['gender'] = df['gender'].replace({'Male':1,'Female':0})
df['OnlineSecurity'] = df['OnlineSecurity'].replace({'Yes':1,'No':0})
df['DeviceProtection'] = df['DeviceProtection'].replace({'Yes':1,'No':0})
df['TechSupport'] = df['TechSupport'].replace({'Yes':1,'No':0})
df['StreamingTV'] = df['StreamingTV'].replace({'Yes':1,'No':0})
df['StreamingMovies'] = df['StreamingMovies'].replace({'Yes':1,'No':0})
df['PaperlessBilling'] = df['PaperlessBilling'].replace({'Yes':1,'No':0})
df['PaymentMethod'] = df['PaymentMethod'].replace({'Electronic check':1,'Mailed check':2,'Bank transfer (automatic)':3,'Credit card (automatic)':4})
df['InternetService'] = df['InternetService'].replace({'No ':0,'Fiber optic':1,'DSL':2})
```

```python
df
```

```python
#Split training and testing set
X=df.drop('Churn',axis=1)
Y=df['Churn']

from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test=train_test_split(X,Y,test_size=0.2,random_state=42)
```

```python
X_train.info()
```

```python
#Covert Columns into int from object
X_train['TotalCharges'] = pd.to_numeric(X_train['TotalCharges'], errors='coerce')
X_test['TotalCharges'] = pd.to_numeric(X_test['TotalCharges'], errors='coerce')

X_train['StreamingMovies'] = pd.to_numeric(X_train['StreamingMovies'], errors='coerce')
X_test['StreamingMovies'] = pd.to_numeric(X_test['StreamingMovies'], errors='coerce')

X_train['StreamingTV'] = pd.to_numeric(X_train['StreamingTV'], errors='coerce')
X_test['StreamingTV'] = pd.to_numeric(X_test['StreamingTV'], errors='coerce')

X_train['TechSupport'] = pd.to_numeric(X_train['TechSupport'], errors='coerce')
X_test['TechSupport'] = pd.to_numeric(X_test['TechSupport'], errors='coerce')

X_train['DeviceProtection'] = pd.to_numeric(X_train['DeviceProtection'], errors='coerce')
X_test['DeviceProtection'] = pd.to_numeric(X_test['DeviceProtection'], errors='coerce')

X_train['OnlineSecurity'] = pd.to_numeric(X_train['OnlineSecurity'], errors='coerce')
X_test['OnlineSecurity'] = pd.to_numeric(X_test['OnlineSecurity'], errors='coerce')

X_train['InternetService'] = pd.to_numeric(X_train['InternetService'], errors='coerce')
X_test['InternetService'] = pd.to_numeric(X_test['InternetService'], errors='coerce')
```

```python
# Replace missing values in the columns with the mean of the column
X_train.fillna({'TotalCharges': X_train['TotalCharges'].mean()}, inplace=True)
X_test.fillna({'TotalCharges': X_test['TotalCharges'].mean()}, inplace=True)

X_train.fillna({'StreamingMovies': X_train['StreamingMovies'].mean()}, inplace=True)
X_test.fillna({'StreamingMovies': X_test['StreamingMovies'].mean()}, inplace=True)

X_train.fillna({'StreamingTV': X_train['StreamingTV'].mean()}, inplace=True)
X_test.fillna({'StreamingTV': X_test['StreamingTV'].mean()}, inplace=True)

X_train.fillna({'TechSupport': X_train['TechSupport'].mean()}, inplace=True)
X_test.fillna({'TechSupport': X_test['TechSupport'].mean()}, inplace=True)

X_train.fillna({'DeviceProtection': X_train['DeviceProtection'].mean()}, inplace=True)
X_test.fillna({'DeviceProtection': X_test['DeviceProtection'].mean()}, inplace=True)

X_train.fillna({'OnlineSecurity': X_train['OnlineSecurity'].mean()}, inplace=True)
X_test.fillna({'OnlineSecurity': X_test['OnlineSecurity'].mean()}, inplace=True)

X_train.fillna({'InternetService': X_train['InternetService'].mean()}, inplace=True)
X_test.fillna({'InternetService': X_test['InternetService'].mean()}, inplace=True)
```

```python
#standardise features
scaler = StandardScaler() 
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

```python
X_test
```

```python
#Logistic Regression
from sklearn.linear_model import LogisticRegression
lg=LogisticRegression()
lg.fit(X_train, Y_train)
Y_pred=lg.predict(X_test)
```

```python
Y_pred
```

```python
#Find Accuracy
from sklearn.metrics import accuracy_score
accuracy_score(Y_test,Y_pred)
```
